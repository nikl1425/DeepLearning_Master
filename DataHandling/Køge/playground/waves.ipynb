```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
dt = 0.001
t = np.arange(0, 1, dt)
f = np.sin(2*np.pi*20*t)
plt.title("Sin wave - 20 Hz")
plt.plot(t, f)
```

```python
f2 = np.sin(2*np.pi*50*t)
plt.plot(t, f2)
plt.title("Sin wave - 50 Hz")
```

```python
f3 = f + f2
plt.plot(t, f3)
plt.title("First & Second Sinus wave")
```

```python
import scipy.fftpack

n = len(f3)
fourier_transform = np.fft.fft(f3, n)
PSD = fourier_transform * np.conj(fourier_transform)
freq = (1/(dt*n)) * np.arange(n)
L = np.arange(1, np.floor(n/2), dtype='int')

plt.plot(freq[L], PSD[L])
plt.xlim(freq[L[0]], 100)
plt.title('Power Spectrum')
plt.legend()
```

```python
from scipy import signal
import matplotlib.pyplot as plt
t = np.linspace(-1, 1, 200, endpoint=False)
sig  = np.cos(2 * np.pi * 7 * t) + signal.gausspulse(t - 0.4, fc=2)
widths = np.arange(1, 31)
cwtmatr = signal.cwt(sig, signal.ricker, widths)
plt.imshow(cwtmatr, extent=[-1, 1, 1, 31], cmap='jet', aspect='auto',
           vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.show()
```
